```python
import numpy as np
import pandas as pd

import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Read results
results = pd.read_csv("results.csv")
results
```

## Varying V, the number of vertices

To show the difference, we plot for certain intervals of E.

### For a graph with 1/3 of max edges

```python
sb.lineplot(data=results[results.e == ((results.v ** 2 - results.v) // 3)], x="v", y="time", hue="type")
```

### For a graph with 1/2 of max edges

```python
sb.lineplot(data=results[results.e == ((results.v ** 2 - results.v) // 2)], x="v", y="time", hue="type")
```

### For a graph with max edges

```python
sb.lineplot(data=results[results.e == ((results.v ** 2 - results.v))], x="v", y="time", hue="type")
```

We can see that for any E, as the number of vertices, V, increases, the time taken for each graph also increases.
This agrees with our theoretical analysis of the time complexity.

Interestingly, you can see that as E increases, the list implementation will take a longer time than the matrix!

## Varying E, the number of edges

To show the difference, we plot for certain intervals of V.

```python
_, axes = plt.subplots(nrows=2, ncols=3, figsize=(30, 30))

sb.lineplot(data=results[results.v == 10], x="e", y="time", hue="type", ax=axes[0][0])
sb.lineplot(data=results[results.v == 20], x="e", y="time", hue="type", ax=axes[0][1])
sb.lineplot(data=results[results.v == 30], x="e", y="time", hue="type", ax=axes[0][2])
sb.lineplot(data=results[results.v == 40], x="e", y="time", hue="type", ax=axes[1][0])
sb.lineplot(data=results[results.v == 50], x="e", y="time", hue="type", ax=axes[1][1])
sb.lineplot(data=results[results.v == 60], x="e", y="time", hue="type", ax=axes[1][2])
```

We can see that for any V, as the number of edges, E, increases, the list implementation will result in worse results than just using a matrix implementation.

This agress with our theoretical analysis of the time complexity for large values of E.
